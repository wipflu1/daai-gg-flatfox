# Imports

```python
import requests
import csv
from dotenv import load_dotenv

load_dotenv()
```

Load data from the API to data

```python
result = requests.get("https://flatfox.ch/api/v1/public-listing/").json()
data = []
data.append(result["results"][0].keys())
for res in result["results"]:
    data.append([res[x] for x in res])
        
while result["next"]:
    result = requests.get(result["next"]).json()
    for res in result["results"]:
        data.append([res[x] for x in res])
    print(result["next"])
print("done")
```

Write data to a csv file

```python
with open("flatfox.csv", mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(data)
```
